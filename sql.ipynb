# Creation of the relationnal database

creation of the date table

```python
min_date= covid_cases.document_date.min()
max_date= covid_cases.document_date.max()
date_table= pd.DataFrame(pd.date_range(min_date,max_date),columns=["date"])
date_table["Date_Id"]=date_table.date.apply(lambda x: str(x).replace("-","")[0:8])
date_table["month"]= date_table["date"].dt.month
date_table["year"]=date_table["date"].dt.year
date_table["Quater"] =date_table["date"].dt.quarter 
date_table["week"]=date_table["date"].dt.isocalendar().week 
date_table
```

creation of department table

```python
departementdf = commune.loc[:, ["adm1code","adm1_en" ]]
departementdf.drop_duplicates(inplace=True)
departementdf.rename(columns={"adm1code":"departement_id", "adm1_en":"departement"}, inplace=True)
departementdf
```

creation of the commune table

```python
commune["population"] = commune["IHSI_UNFPA_2019_female"]+commune["IHSI_UNFPA_2019_male"]
communedf = commune.loc[:, ["adm2code","adm2_en","adm1code","population"]]
communedf.rename(columns={"adm2code":"commune_id","adm1code":"departement_id", "adm2_en":"commune"},inplace=True)
communedf
```

```python
pd.pivot_table(data=communedf, index= "departement_id", values='population', aggfunc='sum')
```

creation of the spa table

```python
spa["depart"] = spa["depart"].apply(lambda x: str(x))
spa["vilcom"] = spa["vilcom"].apply(lambda x: str(x))
spa["commune_id"] = spa["depart"].apply(lambda x: "HT" + x if len(x)==2 else "HT0"+x)
spa["vilcom"] = spa["vilcom"].apply(lambda x: x if len(x)==2 else "0"+x)
spa["commune_id"]= spa["commune_id"]+spa["vilcom"]
spadf=spa.loc[:,["index", "commune_id","facdesc_1","facdesc","mga","service_laboratory","num_beds","ambulance"]]
spadf
```

```python
map_dict1 = {"mga": [1,2,3,4], "mga_type":[ 'public', 'private non profit', 'private with profit', 'private and public']}
mgadf = pd.DataFrame(map_dict1)
mgadf
```

creation of the covid cases table

```python
covid_cases.departement.unique()
```

```python
"""covid_casesdf = covid_cases[-(covid_cases.departement == "Grand Total ")]
covid_casesdf = covid_casesdf[-(covid_casesdf.departement == "Grand Total")]
covid_casesdf = covid_casesdf[-(covid_casesdf.departement == "Guest")]"""

departdf = departement.loc[:,["adm1code", "adm1_fr"]]
departdf.rename(columns={"adm1_fr":"departement"}, inplace=True)
covid_casesdf = pd.merge(left=covid_cases, right=departdf, on="departement", how="left")

covid_casesdf=covid_casesdf[-(covid_casesdf.adm1code.isnull())]

covid_casesdf["Date_Id"] = covid_casesdf.document_date.apply(lambda x: str(x).replace("-","")[0:8])

covid_casesdf.drop(["Unnamed: 0","departement","document_date"], axis=1, inplace=True)
covid_casesdf.rename(columns={"adm1code":"departement_id"}, inplace = 1)
covid_casesdf
```

Connection to sql

```python
import sqlalchemy as sq
```

```python
engine=sq.create_engine("postgresql://postgres:ident@localhost:5432/postgres")
engine.connect
```

```python
spadf.to_sql(name="spa", con=engine, if_exists="replace")
communedf.to_sql(name="commune", con=engine, if_exists="replace")
date_table.to_sql(name="date_table", con=engine, if_exists="replace")
covid_casesdf.to_sql(name="covid", con=engine, if_exists="replace")
mgadf.to_sql(name="mga", con=engine, if_exists="replace")
departementdf.to_sql(name="departement", con=engine, if_exists="replace")
```
